```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('Ferrari.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the original image


scaled_image=cv2.resize(image,(1024 , int(1024*image.shape[0]/image.shape[1])))
plt.imshow(scaled_image)
plt.title('scaled Image')
plt.show()

mask = np.zeros(image.shape[:2],np.uint8)
rect=(50,20,image.shape[1]-100 , image.shape[0]-100)
bgd_model=np.zeros((1,65),np.float64)
fgd_model =np.zeros((1,65),np.float64)

cv2.grabCut(image, mask,rect,bgd_model,fgd_model,2,cv2.GC_INIT_WITH_RECT )


mask2= np.where((mask==2)| (mask==0),0,1).astype('uint8')
output = image*mask2[:,:,np.newaxis]

plt.imshow(output)
plt.axis('off')
plt.show()
```
